```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
```

Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

# Adding title and labels
plt.title('Sine Wave')
plt.xlabel('X Axis (0 to 10)')
plt.ylabel('sin(x)')



plt.show()
```

Multiple lines

```python
x = np.linspace(0, 10, 100)

# Calculating the sine and cosine of these values
y_sin = np.sin(x)
y_cos = np.cos(x)

# Creating line plots for both sine and cosine
plt.plot(x, y_sin, label='sin(x)') # color, linestyle
plt.plot(x, y_cos, label='cos(x)')



# Displaying the plot
plt.show()
```

```python
# Adding title and labels
plt.title('Sine and Cosine Waves')
plt.xlabel('X Axis (0 to 10)')
plt.ylabel('Function Values')

# Adding a legend
plt.legend()
plt.show()
```

Multiple subplots

```python
x = np.linspace(0, 10, 100)

# Creating a figure and a set of subplots
fig, (ax1, ax2) = plt.subplots(1, 2)  # 2 rows, 1 column

# Plotting sine function on the first subplot
ax1.plot(x, np.sin(x))
ax1.set_title('Sine Wave')  # Title for the first subplot
ax1.set_ylabel('sin(x)')    # Y label for the first subplot

# Plotting cosine function on the second subplot
ax2.plot(x, np.cos(x))
ax2.set_title('Cosine Wave') # Title for the second subplot
ax2.set_xlabel('X Axis (0 to 10)') # X label for the second subplot
ax2.set_ylabel('cos(x)')     # Y label for the second subplot

# Adjust the layout
plt.tight_layout()

# Display the figure containing the subplots
plt.show()
```

Bar Plot

```python
categories = ['Category A', 'Category B', 'Category C']
values = [23, 45, 56]

# Creating a bar plot
plt.bar(categories, values)

# Adding title and labels
plt.title('Basic Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Displaying the plot
plt.show()
```

```python
categories = ['Category A', 'Category B', 'Category C']
values = [23, 45, 56]

# Creating a bar plot
plt.bar(categories, values, color=['red', 'green', 'blue'])  # Custom colors

# Adding title and labels
plt.title('Enhanced Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Customizing the grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Displaying the plot
plt.show()
```

```python
categories = ['Category A', 'Category B', 'Category C']
values = [23, 45, 56]

# Creating a horizontal bar plot
plt.barh(categories, values, color=['purple', 'orange', 'skyblue'])

# Adding title and labels
plt.title('Horizontal Bar Plot')
plt.xlabel('Values')
plt.ylabel('Categories')

# Optional: Add value labels to each bar
for index, value in enumerate(values):
    plt.text(value, index, f' {value}')

# Displaying the plot
plt.show()
```

Grouped Bar Chart

```python
categories = ['Category A', 'Category B', 'Category C']
group_a = [23, 45, 56]
group_b = [33, 44, 67]

# X locations for the groups
x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

# Creating the bar plots
plt.bar(x - width/2, group_a, width, label='Group A')
plt.bar(x + width/2, group_b, width, label='Group B')

# Adding title and labels
plt.title('Grouped Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(x, categories)  # Set the x-ticks to match the categories
plt.legend()

# Displaying the plot
plt.show()
```

Stacked Bar Chart

```python
categories = ['Category A', 'Category B', 'Category C']
group_a = [23, 45, 56]
group_b = [33, 44, 67]

# X locations for the groups
x = np.arange(len(categories))

# Creating the bar plots
plt.bar(x, group_a, label='Group A')
plt.bar(x, group_b, bottom=group_a, label='Group B')  # Stacking group_b on top of group_a

# Adding title and labels
plt.title('Stacked Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(x, categories)  # Set the x-ticks to match the categories
plt.legend()

# Displaying the plot
plt.show()
```

Scatter Plot

Triangle, Circle, Square Markers

Box Plots

```python
data = np.random.rand(50)

# Creating a box plot
plt.boxplot(data)

# Adding title and labels
plt.title('Basic Box Plot')
plt.ylabel('Values')

# Displaying the plot
plt.show()
```

```python
data1 = np.random.rand(50)
data2 = np.random.normal(loc=0.5, scale=0.2, size=50)  # Normally distributed data
data3 = np.random.rand(50)

# Creating box plots for multiple datasets
plt.boxplot([data1, data2, data3])

# Adding title and labels
plt.title('Box Plot for Multiple Datasets')
plt.ylabel('Values')
plt.xticks([1, 2, 3], ['Dataset 1', 'Dataset 2', 'Dataset 3'])

# Displaying the plot
plt.show()
```

Violin Plot

Violin plots are similar to box plots, except that they also show the probability density of the data at different values.

```python
data = np.random.normal(loc=0, scale=1, size=100)

# Creating a violin plot
plt.violinplot(data,showmedians=True)

# Adding title and labels
plt.title('Basic Violin Plot')
plt.ylabel('Values')

# Displaying the plot
plt.show()
```

```python
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=2, scale=0.5, size=100)
data3 = np.random.normal(loc=-1, scale=0.8, size=100)

# Creating violin plots for multiple datasets
plt.violinplot([data1, data2, data3], showmedians =True)

# Adding title and labels
plt.title('Violin Plot for Multiple Datasets')
plt.ylabel('Values')
plt.xticks([1, 2, 3], ['Dataset 1', 'Dataset 2', 'Dataset 3'])

# Displaying the plot
plt.show()
```

Histograms

```python
data = np.random.normal(loc=0, scale=1, size=1000)  # Normally distributed data

# Creating a histogram
plt.hist(data, bins=10)  # 10 bins

# Adding title and labels
plt.title('Basic Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Displaying the plot
plt.show()
```

```python
data2 = np.random.normal(loc=2, scale=1.5, size=1000)

# Creating histograms for both datasets
plt.hist(data, bins=20, alpha=0.5, label='Dataset 1')
plt.hist(data2, bins=20, alpha=0.5, label='Dataset 2')

# Adding title and labels
plt.title('Histograms for Multiple Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Displaying the plot
plt.show()
```

```python
import sklearn
import pandas as pd


from sklearn import datasets

# Load the iris dataset
iris = datasets.load_iris()

# Load it to a pandas dataframe:
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['label'])

iris_df['label_names'] = iris_df['label'].apply(lambda x: iris.target_names[int(x)])

# Let's have a look at how the data looks like
iris_df
```

```python
iris_df["label_names"].value_counts()
```

```python
colours = ['orange', 'blue', 'green']
species = iris.target_names

f = plt.figure(figsize=(8,6))

for i in range(0, 3):
    species_df = iris_df[iris_df['label'] == i]
    plt.scatter(
        species_df['petal length (cm)'],
        species_df['sepal length (cm)'],
        color=colours[i],
        alpha=0.5,
        label=species[i]
    )
plt.xlabel('petal length (cm)')
plt.ylabel('sepal length (cm)')
plt.title('Iris dataset: petal length vs sepal length')
plt.legend(loc='lower right')
plt.show()
```
